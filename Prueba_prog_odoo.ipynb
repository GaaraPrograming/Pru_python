```python
# EXAMEN PROG. PYTHON
!pip install qrcode[pil]

import base64
import io
import qrcode
from datetime import datetime

# Punto 1: Agregar campo "Idioma" a listado de clientes
class Cliente:
    def __init__(self, nombre, direccion, telefono, idioma):
        self.nombre = nombre
        self.direccion = direccion
        self.telefono = telefono
        self.idioma = idioma

clientes = [
    Cliente("Agrolait", "Bélgica", "+32 10 588 558", "fr"),
    Cliente("Camptocamp", "Francia", "+33 4 49 23 44 54", "en"),
]

print("🧾 Listado de clientes con idioma:")
for c in clientes:
    print(f"{c.nombre} | {c.direccion} | {c.telefono} | Idioma: {c.idioma}")

# Punto 2: Validación producto con precio 0
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def validar_precio(self):
        if self.precio == 0:
            return f"⚠️ Alerta: El producto '{self.nombre}' tiene precio S/ 0.00"
        return f"✅ Producto '{self.nombre}' agregado correctamente."

producto_test = Producto("Producto Test", 0)
print("\n", producto_test.validar_precio())

# Punto 3: Botón "Boleta" y ventana con total
class Orden:
    def __init__(self, total):
        self.total = total

    def mostrar_boleta(self):
        return f"🧾 Total a pagar: S/ {self.total:.2f}"

orden = Orden(120.50)
print("\n📌 Simulación de botón Boleta:")
print(orden.mostrar_boleta())

# Punto 4: Generar código QR para factura
def generate_qr_code(qr_string):
    qr = qrcode.QRCode(version=4, box_size=4, border=1)
    qr.add_data(qr_string)
    qr.make(fit=True)
    img = qr.make_image()
    buffer = io.BytesIO()
    img.save(buffer, format="PNG")
    img_str = base64.b64encode(buffer.getvalue())
    return img_str

class Factura:
    def __init__(self, numero, cliente, fecha, cantidades_linea, total):
        self.numero = numero
        self.cliente = cliente
        self.fecha = fecha
        self.cantidades_linea = cantidades_linea
        self.total = total

    def generar_qr(self):
        qr_text = f"{self.numero}|{self.cliente}|{self.fecha}|{self.cantidades_linea}|{self.total}"
        return generate_qr_code(qr_text)

factura = Factura("FV/2025/001", "Jeyner Suarez", "2025-07-10", 5, 200.00)
qr_base64 = factura.generar_qr()
print("\n🧾 QR generado (base64):", qr_base64[:50], "...")

# Punto 5: Número de serie y correlativo
def extraer_serie_y_correlativo(numero_factura):
    partes = numero_factura.split("/")
    serie = partes[0] + partes[1]
    correlativo = partes[2].zfill(8)
    return serie, correlativo

serie, correlativo = extraer_serie_y_correlativo("FV/2019/1")
print("\n🆔 Serie:", serie)
print("🔢 Correlativo:", correlativo)

# Punto 6: Canal de ventas cargado por defecto
class CanalVentas:
    def __init__(self, nombre):
        self.nombre = nombre

canales = [CanalVentas("Web"), CanalVentas("Tienda"), CanalVentas("Mayorista")]
print("\n📦 Canales de venta precargados:")
for canal in canales:
    print("-", canal.nombre)

# Punto 7: Reemplazar campo Fecha por Fecha de Emisión
fecha_emision = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print("\n📅 Fecha de emisión actual:", fecha_emision)

# Punto 8: Mostrar en reporte (simulado por consola)
print("\n📄 Reporte de Factura")
print("Serie:", serie)
print("Correlativo:", correlativo)
print("Canal de Venta:", canales[0].nombre)
print("Fecha de Emisión:", fecha_emision)

# Punto 9: Relación de transferencias con pedido de venta
class Transferencia:
    def __init__(self, codigo):
        self.codigo = codigo

class PedidoVenta:
    def __init__(self, id, transferencias):
        self.id = id
        self.transferencias = transferencias

class FacturaRelacionada:
    def __init__(self, pedido):
        self.transferencias = pedido.transferencias

transferencias = [Transferencia("ST/001"), Transferencia("ST/002")]
pedido = PedidoVenta("SO001", transferencias)
factura_rel = FacturaRelacionada(pedido)

print("\n🚚 Transferencias relacionadas a la factura:")
for t in factura_rel.transferencias:
    print("-", t.codigo)
```
